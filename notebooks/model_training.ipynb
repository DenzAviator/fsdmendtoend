```python
# define the customranking for each ordinal variable
cut_categories=["Fair","Good","Very Good","Premium","Ideal"]
color_categories=["D","E","F","G","H","I","J"]
clarity_categories=["I1","SI2", "SI1", "VS2", "VS1", "VVS2","VVS1","IF"]
```

```python
from sklearn.impute import SimpleImputer ## Handling missing values
from sklearn.preprocessing import StandardScaler ##Handling Feature Scaling
from sklearn.preprocessing import OrdinalEncoder ##Ordinal Encoding
##pipelines
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
num_pipeline=Pipeline(
    
    
    steps=[
        
        ("imputer",SimpleImputer()),
        ("scaler",StandardScaler())
        
    ]
)
```

```python
cat_pipeline=Pipeline(
    
    
    steps=[
        
        ("imputer",SimpleImputer(strategy="most_frequent")),
        ("ordinaencoder",OrdinalEncoder(categories=[cut_categories,color_categories,clarity_categories]))
    ]
)
```

```python
preprocessor=ColumnTransformer(
    
    
    [
        ("num_pipeline", num_pipeline,numerical_col),
        ("cat_pipeline", cat_pipeline,categorical_col)
    ]
)
```

```python
## Train test split
```

```python
x_train,x_test,y_tain,_y_test=train_test_split(x,y,test_size=0.30,random_state=30)
```
